#Importing the modules

```python
import random
import pickle

import numpy as np
import pandas as pd
from nltk.tokenize import RegexpTokenizer

from tensorflow.keras.models import Sequential , load_model
from tensorflow.keras.layers import LSTM , Dense, Activation
from tensorflow.keras.optimizers import RMSprop
```

```python
textdf = pd.read_csv("sample_data/fake_or_real_news.csv")
```

```python
textdf
```

# Extracting text and creating tokens



```python
text = list(textdf.text.values)
joined_text = " ".join(text)
```

```python
len(joined_text)
```

```python
par_txt = joined_text[:200000]
```

```python
tokenizer = RegexpTokenizer(r"\w+")
tokens = tokenizer.tokenize(par_txt.lower())
```

```python
tokens
```

```python
unique_tokens = np.unique(tokens)
```

```python
unique_token_index = {token: idx for idx,token in enumerate(unique_tokens)}
```

```python
unique_token_index
```

#Creating the bow(bag of words) Structure

```python
n_words=10
inp_words=[]
nxt_words=[]
```

```python
for i in range(len(tokens)-n_words):
    inp_words.append(tokens[i:i+n_words])
    nxt_words.append(tokens[i+n_words])
```

```python
inp_words
```

```python
X = np.zeros((len(inp_words), n_words, len(unique_tokens)), dtype=bool)
y = np.zeros((len(nxt_words), len(unique_tokens)), dtype=bool)
```

```python
y
```

```python
for i , words in enumerate(inp_words):
    for j, word in enumerate(words):
        X[i,j , unique_token_index[word]]=1
    y[i, unique_token_index[nxt_words[i]]]=1
```

#Defining and Traning the model

```python
model = Sequential()
model.add(LSTM(128, input_shape=(n_words, len(unique_tokens)), return_sequences=True))
model.add(LSTM(128))
model.add(Dense(len(unique_tokens)))
model.add(Activation("softmax"))
```

```python
model.compile(loss="categorical_crossentropy", optimizer=RMSprop(learning_rate=0.01),metrics=["accuracy"])
model.fit(X,y, batch_size=128, epochs=30, shuffle=True)
```

#Saving the Trained Model

```python
model.save("mrk200k.h5")
```

#Loading the Trained model

```python
model = load_model("sample_data/mrk100k.h5")
```

#Next word prediction

```python
def predict_nxt(inp_txt, n_best):
    inp_txt = inp_txt.lower()
    X = np.zeros((1,n_words, len(unique_tokens)))
    for i, word in enumerate(inp_txt.split()):
        X[0,i,unique_token_index[word]]=1
    predictions = model.predict(X)[0]
    return np.argpartition(predictions, -n_best)[-n_best:]
```

#Text Generation

```python
def generate_text(input_text,text_length, creativity =3):
    word_sequence = input_text.split()
    current = 0
    for _ in range(text_length) :
        sub_sequence = " ".join(tokenizer.tokenize(" ".join(word_sequence).lower())[current:current+n_words])
        try:
            choice = unique_tokens[random.choice(predict_nxt(sub_sequence, creativity))]
        except:
            choice = random.choice(unique_tokens)
        word_sequence.append(choice)
        current +=1
    return " ".join(word_sequence)
```
